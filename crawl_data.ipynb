```python
import requests
import csv

# Set the URL for the API
url = "http://universities.hipolabs.com/search?country=United+States"

# Make the GET request
response = requests.get(url)

# Check for a successful response
if response.status_code == 200:
    data = response.json()  # Parse the JSON response

    # Specify the CSV file path
    csv_file_path = 'universities_us.csv'

    # Write to CSV
    with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)

        # Write the header
        writer.writerow(['name', 'alpha_two_code', 'web_page', 'country'])

        # Write each university's data
        for university in data:
            writer.writerow([
                university.get('name'),
                university.get('alpha_two_code'),
                university.get('web_pages')[0] if university.get('web_pages') else '',
                university.get('country')
            ])

    print(f'Data has been written to {csv_file_path}')
else:
    print(f'Error: {response.status_code} - {response.text}')
```
